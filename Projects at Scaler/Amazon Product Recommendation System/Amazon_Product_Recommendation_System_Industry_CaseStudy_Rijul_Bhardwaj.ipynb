### **Section A:**
*Data Understanding & Cleaning: Understand the dataset like a data scientist in a product analytics team.*




```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv("amazon.csv")
df.head()
```

```python

print("Number of unique users:", df['user_id'].nunique())
print("Number of unique products:", df['product_id'].nunique())
print("Number of unique reviews:", df['review_id'].nunique())
```

```python

df['top_category'] = df['category'].apply(lambda x: x.split('|')[0] if pd.notnull(x) else "Unknown")
top_categories = df['top_category'].value_counts().head(5)
print("Top 5 Categories by number of products:\n", top_categories)
```

```python
def extract_price(val):
    if isinstance(val, str):
        return float(val.replace('₹', '').replace(',', '').strip())
    return np.nan

df['discounted_price'] = df['discounted_price'].apply(extract_price)
df['actual_price'] = df['actual_price'].apply(extract_price)
df['discount_percentage'] = df['discount_percentage'].str.replace('%', '').astype(float)

# Check price range
print("Discounted price range:", df['discounted_price'].min(), "-", df['discounted_price'].max())
print("Actual price range:", df['actual_price'].min(), "-", df['actual_price'].max())
```

```python

category_split = df['category'].str.split('|', expand=True)


df['cat_level_1'] = category_split[0]
df['cat_level_2'] = category_split[1]
df['cat_level_3'] = category_split[2]


df[['category', 'cat_level_1', 'cat_level_2', 'cat_level_3']].head()
```

```python
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
df['rating_count'] = df['rating_count'].str.replace(',', '').astype(float)

# Detect outliers in rating
sns.boxplot(df['rating'].dropna())
plt.title('Rating Distribution with Outliers')
plt.show()

df['rating'] = df['rating'].clip(lower=1.0, upper=5.0)
```

```python

df['price_difference'] = df['actual_price'] - df['discounted_price']
df['value_for_money_score'] = df['rating'] / (df['discounted_price'] + 1)
df['weighted_rating'] = df['rating'] * np.log1p(df['rating_count'])
df[['price_difference', 'value_for_money_score', 'weighted_rating']].describe()
```

```python

missing_summary = df.isnull().sum()
print("Missing Values:\n", missing_summary)


df['rating'].fillna(df['rating'].median(), inplace=True)
df['rating_count'].fillna(0, inplace=True)
df.dropna(subset=['discounted_price', 'actual_price'], inplace=True)
```

```python

df.drop_duplicates(inplace=True)

# we beed to Remove invalid price records
df = df[(df['actual_price'] >= df['discounted_price']) & (df['discounted_price'] > 0)]
```

```python

df['review_title'].fillna("No Title", inplace=True)
df['review_content'].fillna("No Content", inplace=True)
```

```python

print("Shape after cleaning:", df.shape)
df.head()
```

# Section C: Content-Based Filtering:

*Act like a content engineer personalizing user feeds based on product metadata.*

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.preprocessing import MinMaxScaler
import numpy as np
```

```python
df['text_features'] = (
    df['product_name'].fillna('') + " " +
    df['about_product'].fillna('') + " " +
    df['category'].fillna('')
)


df['discounted_price'] = df['discounted_price'].fillna(0)
df['actual_price'] = df['actual_price'].fillna(0)
df['discount_percentage'] = df['discount_percentage'].fillna(0)
```

```python
#vectorizing

df['text_features'] = (
    df['product_name'].fillna('') + " " +
    df['about_product'].fillna('') + " " +
    df['category'].fillna('')
)

# Step 1.2: Fill missing values for numeric features
df['discounted_price'] = df['discounted_price'].fillna(0)
df['actual_price'] = df['actual_price'].fillna(0)
df['discount_percentage'] = df['discount_percentage'].fillna(0)


tfidf = TfidfVectorizer(stop_words='english', max_features=5000)
tfidf_matrix = tfidf.fit_transform(df['text_features'])
```

```python
#addingg features

# Normalize numeric features
scaler = MinMaxScaler()
numeric_features = scaler.fit_transform(df[['actual_price', 'discounted_price', 'discount_percentage']])

# Combine text vector with numeric features
from scipy.sparse import hstack
content_matrix = hstack([tfidf_matrix, numeric_features])
```

```python
from sklearn.preprocessing import MinMaxScaler
from scipy.sparse import hstack

# Normalize numeric features
scaler = MinMaxScaler()
numeric_features = scaler.fit_transform(df[['actual_price', 'discounted_price', 'discount_percentage']].fillna(0))

# Combine sparse matrix (TF-IDF) with numeric features
content_matrix = hstack([tfidf_matrix, numeric_features])
```

```python
# now we make the Cosine similarity matrix
similarity_matrix = cosine_similarity(content_matrix)
```

```python
def get_similar_products(product_index, top_n=5):
    sim_scores = list(enumerate(similarity_matrix[product_index]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    top_similar = [i for i, score in sim_scores[1:top_n+1]]
    return df.iloc[top_similar][['product_name', 'rating', 'discounted_price']]
```

```python
# now we find a product with 0 or very low rating_count
new_product_idx = df[df['rating_count'] < 10].index[0]
print("Selected new product:", df.iloc[new_product_idx]['product_name'])

get_similar_products(new_product_idx)
```

```python
# A aproduct with rating < 2.5
low_rating_idx = df[df['rating'] < 2.5].index[0]
print("Low-rated product:", df.iloc[low_rating_idx]['product_name'])

get_similar_products(low_rating_idx)
```

```python

similarities = similarity_matrix[new_product_idx][1:]
print("Average similarity score:", np.mean(similarities))
print("Max similarity score:", np.max(similarities))

# You can also measure average category overlap for top recommendations
def category_overlap(index):
    target_cat = df.iloc[index]['cat_level_1']
    similar = get_similar_products(index)
    return similar['product_name'].count(), similar[df['cat_level_1'] == target_cat].shape[0]

category_overlap(new_product_idx)
```

# **Section D: Collaborative Filtering (User–Item) :**
*Now you're a machine learning engineer building smart recommendations using user behavior.*

```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
```

```python

df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
df = df.dropna(subset=['user_id', 'product_id', 'rating'])

# creating User-Item Rating Matrix (pivot table) - refer to shivank sir class
user_item_matrix = df.pivot_table(index='user_id', columns='product_id', values='rating')
user_item_matrix.fillna(0, inplace=True)
user_item_matrix.head()
```

```python
# USER-USER Collaborative Filtering using cosine similarity
user_similarity = cosine_similarity(user_item_matrix)
user_similarity_df = pd.DataFrame(user_similarity, index=user_item_matrix.index, columns=user_item_matrix.index)

# OR we can use ITEM-ITEM
# item_similarity = cosine_similarity(user_item_matrix.T)
# item_similarity_df = pd.DataFrame(item_similarity, index=user_item_matrix.columns, columns=user_item_matrix.columns)
```

```python

def recommend_products(user_id, top_n=5):
    if user_id not in user_item_matrix.index:
        return f"User {user_id} not found."


    sim_scores = user_similarity_df[user_id]


    weighted_ratings = user_item_matrix.T.dot(sim_scores)
    sim_sums = np.array([sim_scores.sum()] * user_item_matrix.shape[1])
    predicted_ratings = weighted_ratings / sim_sums
    predicted_ratings = pd.Series(predicted_ratings, index=user_item_matrix.columns)


    already_rated = user_item_matrix.loc[user_id]
    unseen_products = predicted_ratings[already_rated == 0]


    recommendations = unseen_products.sort_values(ascending=False).head(top_n)
    return df[df['product_id'].isin(recommendations.index)][['product_name', 'product_id']].drop_duplicates()
```

```python

sample_user_id = user_item_matrix.index[0]
print(f"Recommendations for user: {sample_user_id}")

recommend_products(sample_user_id)
```

# Section E: Hybrid Recommender (Content + Collaborative) :
*Step into the role of a senior ML engineer combining models for better performance.*

```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.preprocessing import MinMaxScaler
from sklearn.feature_extraction.text import TfidfVectorizer
from scipy.sparse import hstack
```

```python

df['text_features'] = (
    df['product_name'].fillna('') + " " +
    df['about_product'].fillna('') + " " +
    df['category'].fillna('')
)

tfidf = TfidfVectorizer(stop_words='english', max_features=5000)
tfidf_matrix = tfidf.fit_transform(df['text_features'])


scaler = MinMaxScaler()
num_features = scaler.fit_transform(df[['actual_price', 'discounted_price', 'discount_percentage']].fillna(0))

content_matrix = hstack([tfidf_matrix, num_features])
content_sim = cosine_similarity(content_matrix)
```

```python
# User–Item matrix
user_item_matrix = df.pivot_table(index='user_id', columns='product_id', values='rating').fillna(0)

# now im using  item–item similarity
cf_sim = cosine_similarity(user_item_matrix.T)
cf_sim_df = pd.DataFrame(cf_sim, index=user_item_matrix.columns, columns=user_item_matrix.columns)
```

```python

common_product_ids = df['product_id'].unique()
product_idx_map = {pid: i for i, pid in enumerate(common_product_ids)}

# Align both similarity matrices (content-based and collaborative)
cf_score_matrix = cf_sim_df.reindex(index=common_product_ids, columns=common_product_ids).fillna(0).values
content_score_matrix = content_sim[:len(common_product_ids), :len(common_product_ids)]

# Fuse the matrices
hybrid_score_matrix = 0.6 * cf_score_matrix + 0.4 * content_score_matrix
```

```python
def hybrid_recommendations(product_id, top_n=5):
    if product_id not in product_idx_map:
        return "Product not found"

    idx = product_idx_map[product_id]
    sim_scores = list(enumerate(hybrid_score_matrix[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)[1:top_n+1]
    indices = [i for i, score in sim_scores]

    return df.iloc[indices][['product_name', 'rating', 'discounted_price']].drop_duplicates()
```

```python
# Find product with few or no ratings
cold_start_product = df[df['rating_count'] < 10].iloc[0]['product_id']
print("Cold start product:", cold_start_product)
hybrid_recommendations(cold_start_product)
```

```python

user_review_counts = user_item_matrix[user_item_matrix > 0].astype(bool).sum(axis=1)


cold_users = user_review_counts[user_review_counts < 3]


if len(cold_users) == 0:
    print("No cold-start users found in this dataset.")
else:

    cold_user = cold_users.index[0]
    rated_products = user_item_matrix.loc[cold_user][user_item_matrix.loc[cold_user] > 0]

    if rated_products.empty:
        print("Cold-start user has not rated any product. Recommend popular or trending items.")
    else:
        rated_product = rated_products.index[0]
        print("Cold-start user rated product:", rated_product)
        hybrid_recommendations(rated_product)
```

```python
print("Individual Content-Based Recommendations:")
print(df.iloc[content_score_matrix[product_idx_map[cold_start_product]].argsort()[-6:-1]][['product_name']])

print("\nIndividual Collaborative Recommendations:")
print(cf_sim_df[cold_start_product].sort_values(ascending=False).iloc[1:6])

print("\nHybrid Recommendations:")
print(hybrid_recommendations(cold_start_product))
```

```python
improvements = [
    "1. Popularity Boost: Add weights based on total review counts or views (e.g., via a popularity factor).",
    "2. Recency Score: Incorporate timestamp decay — newer products get a freshness boost.",
    "3. Availability Check: Filter out out-of-stock or discontinued products during recommendation.",
    "4. Personalized Re-ranking: Blend with user purchase history, browsing events, or demographic clustering.",
    "5. Multi-objective Optimization: Factor in business goals like margin, seller rating, or delivery speed."
]

print("Ways to Improve Hybrid Recommender:")
for tip in improvements:
    print(tip)
```

##  Section F: Business Strategy & Deployment

---

###  1. Which model works best for new users?

For **new users** (who haven’t rated or viewed many products), the best approach is **Content-Based Filtering**.  
It doesn’t rely on past interactions — instead, it recommends products based on product metadata like name, description, category, price, and discount.

**How to improve it further:**
- Use onboarding questions (preferred categories, budget range, etc.)
- Blend in popular or trending items for better discovery

---

###  2. Which model works best for returning users?

For **returning users** who have interacted with multiple products, **Collaborative Filtering** is more effective.  
It finds similar users and recommends products they liked — making it more personalized and relevant.

**Best practice:** Use a **Hybrid Model** that combines both collaborative and content-based methods for balanced recommendations.

---

###  3. How can we recommend products with no ratings?

When a product is new and has no reviews or ratings, we can still recommend it using:

- **Content-Based Filtering**: Match new products with existing ones using metadata (name, category, description, price).
- **Popularity & recency boosting**: Promote new arrivals, limited-time offers, or brand-new launches.
- **Fallback strategies**: Use brand popularity, editor’s picks, or trending product clusters.

---

###  4. How would you deploy this system in production?

A production-grade recommendation system must be fast, scalable, and responsive to real-time updates.

**Core components and tools:**

| Component               | Tools/Technologies                                       |

- Pre-compute similarity matrices daily via batch jobs.
- Store recommendations in a cache (e.g. Redis).
- Serve via API to website or app based on user context.
